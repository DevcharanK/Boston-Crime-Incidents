```python
import pandas as pd

# Read the Excel file
file_path = 'External_Factors.xlsx'
sheet_name = 'Household Income'
df = pd.read_excel(file_path, sheet_name)

# Display the basic statistics
statistics = df.describe()
print(statistics)

# Calculate the median income for each location
median_income_by_location = df.groupby('DISTRICT')['Median Income'].median()
print(median_income_by_location)
```

```python
crime_df = pd.read_csv('Processed_data.csv')
```

```python
# Calculate crime rates by district
crime_rates_by_district = crime_df['DISTRICT'].value_counts().reset_index()
crime_rates_by_district.columns = ['DISTRICT', 'CrimeRate']
```

```python
# Create a mapping between district names and codes
district_mapping = {
    'Brighton D14': 'D14',
    'Charlestown A15': 'A15',
    'Dorchester C11': 'C11',
    'Downtown A1': 'A1',
    'East Boston A7': 'East Boston',
    'Hyde Park E18': 'E18',
    'Jamaica Plain E13': 'E13',
    'Mattapan B3': 'B3',
    'Roxbury B2': 'B2',
    'South Boston C6': 'C6',
    'South End D4': 'D4',
    'West Roxbury E5': 'E5',
}
```

```python
# Map the district names in income data to codes
df['DistrictCode'] = df['DISTRICT'].map(district_mapping)
```

```python
# Calculate crime rates by district
crime_rates_by_district = crime_df['DISTRICT'].value_counts().reset_index()
```

```python
crime_rates_by_district.columns = ['DistrictCode', 'CrimeRate']
```

```python
# Merge with income data
merged_df = pd.merge(crime_rates_by_district, df, on='DistrictCode', how='inner')
```

```python
# Calculate correlation matrix
correlation_matrix = merged_df[['Median Income', 'CrimeRate']].corr()
```

```python
# Display the crime rates by district
print(crime_rates_by_district)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualize the distribution of income across different areas
plt.figure(figsize=(12, 6))
sns.boxplot(x='DISTRICT', y='Median Income', data=merged_df, order=merged_df.groupby('DISTRICT')['Median Income'].median().sort_values().index)
plt.xticks(rotation=90)
plt.title('Distribution of Median Income Across Different Areas')
plt.show()

# Visualize crime rates across different areas
plt.figure(figsize=(12, 6))
sns.boxplot(x='DISTRICT', y='CrimeRate', data=merged_df, order=merged_df.groupby('DISTRICT')['CrimeRate'].median().sort_values().index)
plt.xticks(rotation=90)
plt.title('Crime Rates Across Different Areas')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot the trend of median income
sns.lineplot(x='DISTRICT', y='Median Income', data=merged_df, label='Median Income', marker='o', ax=ax1)
ax1.set_xlabel('District')
ax1.set_ylabel('Median Income', color='blue')
ax1.tick_params('y', colors='blue')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

# Create a second y-axis to plot the trend of crime rates
ax2 = ax1.twinx()
sns.lineplot(x='DISTRICT', y='CrimeRate', data=merged_df, label='Crime Rate', marker='s', ax=ax2, color='orange')
ax2.set_ylabel('Crime Rate', color='orange')
ax2.tick_params('y', colors='orange')

# Add legend
ax1.legend(loc='upper left', bbox_to_anchor=(0.7, 1))

# Title
plt.title('Comparison of Median Income and Crime Rates Across Different Areas')

# Show the plot
plt.show()
```
